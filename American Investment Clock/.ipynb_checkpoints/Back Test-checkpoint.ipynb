```python
[change,circle_type]= Circle_Identify.Circle_Identify(china)# change: record the index of circle change in the china; circle_type: record the circle type
price# assets price for the back test
day= china['日期']
day=pd.DataFrame(day)
day=day['日期'].map(lambda a : a.strftime('%Y-%m')) 
price_date=price.index.map(lambda a: a.strftime('%Y-%m'))
price.index=price_date #因为是月度数据，有时候日对不上，所以仅提取年和月
price
```

<font face='黑体' size= 4 color=#8B0000 >回测实现 </font>

```python
testdate=[]#record the asset changing day
r=[]#record the return.
changedate=[]
for nbr in  range(len(change)): #这里其实是不包括最后一次的调仓的
    if day[change[nbr]] in price.index:
        circletype=circle_type[nbr]
        changedate.append(day[change[nbr]])
        if (nbr< len(change)-1):
            res=period_return(day[change[nbr]],day[change[nbr+1]],circletype,price)
            temp_r=res[0]
            temp_d=res[1]
        else:
            res=period_return(day[change[nbr]],price.index[-1],circletype,price)
            temp_r=res[0]
            temp_d=res[1]
        r=r+temp_r #直接将收益率列表合并了
        testdate=testdate+temp_d  
    else:
        ;
    nbr=nbr+1 
```

```python
changedate
testdate
plt.plot(r)
```

```python
d=np.array(r)+1
nav=np.cumprod(d)
res=pd.DataFrame(nav,index=testdate,columns=['净值'])
import datetime
start_date=changedate[0]
res.loc[start_date]=1
res=res.sort_index()
res
```

<font face='黑体' size= 4 color=#8B0000 >相关指标 </font>

```python
whole_return=res.values
annual_return=(whole_return[-1])**(254/len(whole_return))-1
print('年化收益率',annual_return)

#计算年化波动率
import math
annual_vol=whole_return.std()*math.sqrt(12)
print('年化波动率',annual_vol)

#计算夏普比率
print('夏普比率',(annual_return/annual_vol))

#计算最大回撤
return_list=whole_return
max_turndown=((np.maximum.accumulate(return_list) - return_list) / np.maximum.accumulate(return_list)).max()
print('最大回撤',max_turndown)

#计算Calmar值
calmar=annual_return/max_turndown
print('Calmar',calmar)
```

```python
def period_return(d1,d2,circle,D):
    #d1,d2是持有资产的起始和终止日期
    if circle==0:
        asset='沪深300'
    elif circle==1:
        asset='南华商品指数'
    elif circle==2:
        asset='货币基金'
    else:
        asset='中债-企业债总财富(总值)指数'
    asset_choose=D[asset][d1:d2]#选中投资资产
    temp_r=asset_choose/asset_choose.shift(1)-1
    temp_r=temp_r.dropna()
    date=list(temp_r.index)
    ret=list(temp_r.values)
    return[ret,date]#ret 从起始日之后的第一期记录，日期也是。
```
