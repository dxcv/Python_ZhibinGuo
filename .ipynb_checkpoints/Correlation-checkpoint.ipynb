```python
import os
import pandas as pd
import numpy as np
import os 
```

```python
def HP_Filter(Y,factor):
    #Y is the original array,factor is the penalty factor
    #Y is a column vector, type is list/array
    n=len(Y)
    if (n<4):
        print("The number of the sample must bigger than 4")
        exit()
    else:
        F=np.zeros([n,n])
        for i in range(0,n):
            if (i==0):
                F[i][0]=1
                F[i][1]=-2
                F[i][2]=1
            elif (i==1):
                F[i][0]=-2
                F[i][1]=5
                F[i][2]=-4
                F[i][3]=1
            elif(1<i<n-2):
                F[i][i-2]=1
                F[i][i-1]=-4
                F[i][i]=6
                F[i][i+1]=-4
                F[i][i+2]=1
            elif(i==n-2):
                F[i][-1]=-2
                F[i][-2]=5
                F[i][-3]=-4
                F[i][-4]=1
            else:
                F[i][-1]=1
                F[i][-2]=-2
                F[i][-3]=1
        F=np.matrix(F)#transfer into matrix
        I=np.matrix(np.identity(n))#generate identity matrix
        M=I+factor*F
        Y=np.matrix(Y)
        G=np.linalg.inv(M)*(Y.T)
        C=Y.T-G
        G=np.array(G)
        C=np.array(C)
    return G,C
```

```python
def CORR(D1,D2,lag):
    #D1,D2 are two DataFrame,D2=GDP
    # backward lag
    if lag>=0:
        V2=D2.shift(lag).dropna()
        V1=D1.iloc[lag:] #V1=GDP
    else: #forward lag
        V2=D2.shift(lag)
        V1=D1.iloc[:lag]# V1=GDP
    return V1.corr(V2)
```

```python
#导入数据
os.chdir("D:\各种材料\研究生\课程\Banking\Homework 1")
data=pd.read_excel(r"D:\各种材料\研究生\课程\Banking\Homework 1\data.xlsx")
data=data.dropna()
data.index=data["Date"]
del data["Date"]
data=np.log(data)
```

```python
#滤波获得各个变量的周期项
name=data.columns
Date=data.index
eco_dict={}#To save the circle dataframe
for i in name:
    eco=data[i].values.tolist()
    [trend,circle]=HP_Filter(eco,1600)
    temp_dict=pd.DataFrame(circle,index=Date)
    eco_dict[i]=temp_dict  
circle_array=[eco_dict[i]   for i in eco_dict.keys()]
circle=pd.concat(circle_array,axis=1)
circle.columns=name
circle
```

```python
#计算标准差和相关系数
STD=[]
correlation=[]
GDP=circle["GDP"]
for i in circle.columns:
    STD.append(circle[i].std())
    corr_temp=[]
    for j in range(-6,7):
        corr_temp.append(CORR(GDP,circle[i],j))
    correlation.append(corr_temp) 
```

```python
res_std=pd.DataFrame(STD,index=name,columns=["std"])
res_corr=pd.DataFrame(correlation,index=name,columns=range(-6,7))
res=pd.concat([res_std,res_corr],axis=1)
res
```
