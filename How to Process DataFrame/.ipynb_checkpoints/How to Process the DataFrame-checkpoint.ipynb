<font size=4 color=#08B0000 face='time new roman'>**以年份为键，以当年所有的交易日为键值建立字典**<font>

```python
riqi={}
for i in range(1999,2020):
    temp=[]
    for T in risk_asset.index:
        if T.year==i:
            temp.append(T)
    riqi[str(i)]=temp
riqi['2006'][:10]
```

<font size=4 color=#08B0000 face='time new roman'>**以年份为键，键值为"以月份为键值，当月所有交易日为列表的的字典"**<font>

```python
year_month_day={}
for i in range(1999,2020):
    date_list=riqi[str(i)]#得到该年份的所有交易日
    months=[]
    for T1 in date_list:
        temp_m=T1.month
        months.append(temp_m)
    months=list(set(months))#得到该年份有交易日的交易月份
    month_date={}
    for month in months:
        temp_day=[]#用来存储每个月的交易日
        for T2 in date_list:
            if T2.month==month:
                temp_day.append(T2)
        month_date[str(month)]=temp_day
    year_month_day[str(i)]=month_date

year_month_day['1999']
```
